```python
# Hitung frekuensi masing-masing isi
salary_counts = gbg1['salary_preview'].value_counts().head(30)  # ambil 20 teratas biar tidak terlalu padat

# Plot sebagai bar chart
plt.figure(figsize=(10,6))
salary_counts.plot(kind='barh')
plt.xlabel('Jumlah')
plt.ylabel('Isi Salary dalam $ (dipotong)')
plt.title('Frekuensi Salary Preview (Top 20)')
plt.gca().invert_yaxis()  # agar ranking tertinggi di atas
plt.tight_layout()
plt.show()
```

```python
# Hitung frekuensi masing-masing isi (top 20)
salary_counts = gbg1['salary_preview'].value_counts().head(20)

# Buat plot
plt.figure(figsize=(10,6))
ax = salary_counts.plot(kind='barh', color='skyblue', edgecolor='black')

# Judul & label
plt.xlabel('Jumlah')
plt.ylabel('Isi Salary (dipotong)')
plt.title('Frekuensi Salary Preview (Top 20)')
plt.gca().invert_yaxis()  # ranking tinggi di atas

# Tambahkan annotation di ujung bar
for i, (value, count) in enumerate(salary_counts.items()):
    plt.text(count + 1, i, str(count), va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

# General Conclusion


- 📊 General Conclusions from JobsDB Singapore Dataset (3,532 entries)



- 🧾 Job Titles
     The most frequently listed job is "Personal Assistant", indicating strong demand in administrative support roles.

    Other popular roles include Customer Service, Marketing Executive, and Data Entry, reflecting demand in operations and communication-focused positions.



- 💼 Top Hiring Companies
   Certain companies such as Superhero Productions, Johnson Controls, and Micron Semiconductors consistently post more job listings.

   These firms may represent high-growth or labor-intensive industries.


- 📍 Top Locations
    The majority of job vacancies are concentrated in:

    Singapore (city-wide listing)

    Central Singapore

    CBD Area

    This shows that urban and central business districts dominate job availability.



- 💰 Salary Insights
    The most frequently mentioned salary range is $4,000–$5,000 per month.

    Salary previews with terms like "Full time" or "Contract" (non-numeric) were categorized as "tbi" (to be informed).

    There are many listings without clear salary data, highlighting a lack of transparency or variable compensation models.



- 💡 Additional Insights
    A significant number of job listings do not include salary information, making it harder for job seekers to assess opportunities.

    There are salary entries with inconsistent formatting (e.g. line breaks), which suggests that data cleaning is essential for accurate analysis.

    Companies with high job posting counts might indicate either fast scaling, high turnover, or multiple open roles across functions.



- ✅ Suggestions for Job Seekers or Analysts:
    Focus on Central Singapore or CBD listings for broader opportunities.

    Personal Assistant, Admin, and Entry-level roles seem widely available.

    Consider targeting companies with frequent postings, but evaluate employee reviews to understand the work environment.
